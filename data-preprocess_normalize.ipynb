```python
from itertools import repeat
import numpy as np
import random
from sklearn import preprocessing


def floatX(X):
    return np.asarray(X, dtype=theano.config.floatX)

def vectorized_result ( j , siz) :
    e = np.zeros((siz, 1))
    e[j] = 1.0
    return np.reshape( e, siz)

TRAINING_DATA   = './dataCebuano/train.f0_ffv_fbank.fea'
VALIDATION_DATA = './dataCebuano/dev.f0_ffv_fbank.fea'

# Dataset
x_data = []
x_name = []

# index for seperate the traning data and validation data
separate = 0 

record_seq = []
with open( TRAINING_DATA, 'r') as f:
    for lines in f: 
        if '[' in lines :
            id = lines.split(' ')[0]
            x_name.append(id)
            seq = 0
        elif ']' in lines :
            seq += 1
            separate += 1
            record_seq.append(seq)
            x_data.append([float(x) for x in lines.split(' ') [2:-1] ])
        else :
            seq += 1
            separate += 1
            x_data.append([float(x) for x in lines.split(' ') [2:-1] ])


with open( VALIDATION_DATA, 'r') as f:
    for lines in f: 
        if '[' in lines :
            id = lines.split(' ')[0]
            x_name.append(id)
            seq = 0
        elif ']' in lines :
            seq += 1
            record_seq.append(seq)
            x_data.append([float(x) for x in lines.split(' ') [2:-1] ])
        else :
            seq += 1
            x_data.append([float(x) for x in lines.split(' ') [2:-1] ])
        
print "Normalized"
x_data = preprocessing.scale( np.array(x_data) )
```
